# Visualization Data with Matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
prices = winemag[winemag.price > 10].price
```

```python
winemag.plot(subplots = True, figsize=(15, 15))
plt.show()
```

```python
winemag.price.plot(subplots = True, figsize=(20, 20))
plt.show()
```

```python
winemag.price.plot(kind = 'hist', figsize=(15, 5), bins = 81)
plt.show()
```

```python
winemag.plot(kind = 'scatter', figsize=(15, 10), x = 'price', y = 'points')
plt.xlabel('Price', fontsize=16)
plt.ylabel('Points', fontsize=16)
plt.title('Winemag price and point', fontsize = 20)
plt.show()
```

```python
winemag.country.value_counts(ascending=True).plot()
plt.show()
```
